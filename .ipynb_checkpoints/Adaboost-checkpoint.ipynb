```python
query = "SELECT * FROM midb_ab_e_econ2017;"
data = pd_sql.read_sql(query, connection)
dum_outcome = pd.get_dummies(data['outcome'], drop_first=True)
dum_settle = pd.get_dummies(data['settle'], drop_first=True)
dum_revision_type_1_a = pd.get_dummies(data['revision_type_1_a'], drop_first=True)
dum_revision_type_1_b = pd.get_dummies(data['revision_type_1_b'], drop_first=True)
data = data.drop(['outcome', 'settle', 'revision_type_1_a', 'revision_type_1_b'], axis=1)
data = pd.concat([data, dum_outcome, dum_settle, dum_revision_type_1_a, dum_revision_type_1_b], axis=1)
data.drop('highest_hostile_action[hostility_level]', axis = 1, inplace=True)
data['duration'] = data['mid_end']-data['mid_start']
data.drop(['mid_end','mid_start'], axis = 1, inplace=True)
data.drop('index', axis = 1, inplace=True)
data.drop('maximum_diration', axis = 1, inplace=True)
data.drop('fatality_bucket', axis = 1, inplace=True)
data.drop('minimum_duration', axis = 1, inplace=True)
```

```python
train = data[data['start_year'] < 1990]
test = data[data['start_year'] > 1990]
y_train = train['hostility_level']#.to_numpy().reshape(1, -1)
y_test = test['hostility_level']
X_train = train.loc[:, train.columns != 'hostility_level']
X_test = test.loc[:, test.columns != 'hostility_level']
```

```python
X = data.loc[:, data.columns != 'hostility_level']
y = data['hostility_level']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.30, random_state=22)
```

#### Scaling


#### Oversampling

```python
n_pos = np.sum(y_train == 1)
n_neg = np.sum(y_train == 0)
ratio = {1: n_pos, 0: n_neg * 2}
ROS = imblearn.over_sampling.RandomOverSampler(sampling_strategy = ratio, random_state=22)
X_train_rs, y_train_rs = ROS.fit_resample(X_train, y_train)
```

### Adaboost

```python
ada_random_oversampling = AdaBoostClassifier(DecisionTreeClassifier(),
                         algorithm="SAMME.R",
                         n_estimators=300, learning_rate=0.8)

ada_random_without_oversampling = AdaBoostClassifier(DecisionTreeClassifier(),
                         algorithm="SAMME.R",
                         n_estimators=300, learning_rate=0.8)

ada_time_oversampling = AdaBoostClassifier(DecisionTreeClassifier(),
                         algorithm="SAMME.R",
                         n_estimators=300, learning_rate=0.8)

ada_time_without_oversampling = AdaBoostClassifier(DecisionTreeClassifier(),
                         algorithm="SAMME.R",
                         n_estimators=300, learning_rate=0.8)
```

```python
def Overall_result(model,X_train,y_train,X_test,y_test):
    print('The score for Adaboost is')
    print('Training: {:6.2f}%'.format(100*model.score(X_train, y_train)))
    print('Test: {:6.2f}%'.format(100*model.score(X_test, y_test)))
    y_pred = model.predict(X_test)
    print('Precision: {:6.4f}, Recall: {:6.4f}'.format(precision_score(y_test, y_pred),recall_score(y_test, y_pred)))
    print('Confusion Matrix: \n\n', confusion_matrix(y_test, model.predict(X_test)))
    

def Confusion_Matrix(model,X_train,y_train,X_test,y_test):
    sns.set(font_scale=2.0, style='white', palette = 'Paired')
    plot_confusion_matrix(model, X_test, y_test, values_format='d', cmap='Blues', display_labels=['Threat', 'War'])
    plt.title('Adaboost Confusion Matrix', x=0.5, y=1.05)
    plt.xlabel('Predicted', fontsize=13)
    plt.ylabel('True', fontsize=13,labelpad=15)
    
def Precision_Recall_Curves(model,X_train,y_train,X_test,y_test):
    precision_curve, recall_curve, threshold_curve = precision_recall_curve(y_test, model.predict_proba(X_test)[:,1])
    plt.figure(dpi=80)
    plt.plot(threshold_curve, precision_curve[1:], label='precision')
    plt.plot(threshold_curve, recall_curve[1:], label='recall')
    plt.xlabel('Threshold (above this probability) label as 1')
    plt.title('Precision and Recall Curves')
    plt.legend(loc='lower right')
    plt.figure(dpi=80)
    plt.plot(recall_curve[1:], precision_curve[1:])
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.title('Precision-Recall Curve')
    
def ROC_AUC_Scores(model,X_train,y_train,X_test,y_test):
    fpr, tpr, thresholds = roc_curve(y_test, model.predict_proba(X_test)[:,1])
    roc_auc = roc_auc_score(y_test, model.predict_proba(X_test)[:,1])
    roc_auc = round(roc_auc, 2)
    plt.plot(fpr,tpr,lw=2)
    plt.plot([0,1],[0,1], ls='--')
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.text(0.6, 0.2, 'ROC AUC Score: %.2f' % roc_auc)
    
    
def get_model(model,X_train,y_train,X_test,y_test,over_sampling):
    
    if over_sampling == True:
        n_pos = np.sum(y_train == 1)
        n_neg = np.sum(y_train == 0)
        ratio = {1: n_pos, 0: n_neg * 2}
        ROS = imblearn.over_sampling.RandomOverSampler(sampling_strategy = ratio, random_state=22)
        X_train_rs, y_train_rs = ROS.fit_resample(X_train, y_train)
        model.fit(X_train_rs, y_train_rs)
    else:
        model.fit(X_train, y_train)   
    
    return model 
    
        
```
